```python
def xgb_classifier(anomaly_data, anomaly_name_list, features_list_to_be_used):
    data_to_be_classified = anomaly_data
    anomaly_list = {3:'flange',
                    11:'valve',
                    12:'weld',
                    2:'attachment',
                    9:'tap start',
                    7:'sleeve start',
                    6:'sleeve end',
                    10:'unknown',
                    1:'area start',
                    0:'area end',
                    5:'metal loss',
                    4:'marker'
                   }
    #Data preparation
    features_list_to_be_used = ['min_of_moving_avg',
                               'max_of_moving_avg', 
                               'mean_of_moving_avg', 
                               'min_of_Max',
                               'max_of_Max',
                               'mean_of_Max',
                               'min_of_Min',
                               'max_of_Min',
                               'mean_of_Min',
                               'min_of_delta',
                               'max_of_delta', 
                               'mean_of_delta', 
                               'min_of_local_std',
                               'max_of_local_std',
                               'mean_of_local_std',
                               'min_of_global_std',
                               'max_of_global_std',
                               'mean_of_global_std',
                               'min_of_idx_delta',
                               'max_of_idx_delta',
                               'mean_of_idx_delta',
                               'min_of_avg', 
                               'max_of_avg',
                               'mean_of_avg',
                               'anomalous_count',
                               'max_span',
                               'anomaly_length',
                               'intensity']

    featured_data = data_to_be_classified[features_list_to_be_used]
    featured_data_arr = featured_data.values
    matrix_data = xgb.DMatrix(featured_data_arr)
    preds = XGBoost.predict(matrix_data)
    best_preds = np.asarray([np.argmax(line) for line in preds])
    anomaly_data['anomaly_code'] = best_preds
    anomaly_data['anomaly_name'] = anomaly_data.apply(lambda row: anomaly_name_list.get(int(row['anomaly_code'])), axis=1)
    return anomaly_data
```
